### Task 1.2 Integrarate Data

```python
# Rename columns in PHN_Hosp_Death to match with final_df for merging
phn_hosp_death_df_renamed = phn_hosp_death_df.rename(columns={
    'Gender': 'Sex',
    'Age': 'Age_Group',
    'Annualised hospitalisations': 'Annualised_Hospitalisations',
    'Hospitalisations per 100,000': 'Hospitalisations_per_100k',
    'Annualised population': 'Annualised_Population_Hosp',
    'Annualised deaths': 'Annualised_Deaths',
    'Deaths per 100,000': 'Deaths_per_100k',
    'Annualised population.1': 'Annualised_Population_Death'
})
```

```python
# Standardize merge key columns in both dataframes
for df in [final_df, phn_hosp_death_df_renamed]:
    df['Code'] = df['Code'].astype(str).str.strip()
    df['Name'] = df['Name'].astype(str).str.strip().str.lower()
    df['Sex'] = df['Sex'].astype(str).str.strip().str.lower()
    df['Age_Group'] = df['Age_Group'].astype(str).str.strip().str.lower()
```

```python
# Merge based on the specified keys
merged_df = pd.merge(
    final_df,
    phn_hosp_death_df_renamed[
        ['Code', 'Name', 'Sex', 'Age_Group',
         'Annualised_Hospitalisations', 'Hospitalisations_per_100k',
         'Annualised_Population_Hosp', 'Annualised_Deaths',
         'Deaths_per_100k', 'Annualised_Population_Death']
    ],
    on=['Code', 'Name', 'Sex', 'Age_Group'],
    how='left'
)
```

```python
merged_df.head()
```

```python
# Rename PHA columns to match merged_df for conditional merging
pha_hosp_df_renamed = pha_hosp_df.rename(columns={
    'Gender': 'Sex',
    'Age': 'Age_Group',
    'Annualised hospitalisations': 'Annualised_Hospitalisations',
    'Hospitalisations per 100,000': 'Hospitalisations_per_100k',
    'Annualised population': 'Annualised_Population_Hosp'
})
```

```python
# Standardize keys in PHA data
for df in [pha_hosp_df_renamed]:
    df['Code'] = df['Code'].astype(str).str.strip()
    df['Name'] = df['Name'].astype(str).str.strip().str.lower()
    df['Sex'] = df['Sex'].astype(str).str.strip().str.lower()
    df['Age_Group'] = df['Age_Group'].astype(str).str.strip().str.lower()
```

```python
# Merge PHA data on same keys but suffix PHA columns temporarily
merged_df = pd.merge(
    merged_df,
    pha_hosp_df_renamed[['Code', 'Name', 'Sex', 'Age_Group',
                         'Annualised_Hospitalisations', 'Hospitalisations_per_100k',
                         'Annualised_Population_Hosp']],
    on=['Code', 'Name', 'Sex', 'Age_Group'],
    how='left',
    suffixes=('', '_PHA')
)
```

```python
# Fill missing values in existing columns using PHA values
for col in ['Annualised_Hospitalisations', 'Hospitalisations_per_100k', 'Annualised_Population_Hosp']:
    merged_df[col] = merged_df[col].combine_first(merged_df[f"{col}_PHA"])
    merged_df.drop(columns=[f"{col}_PHA"], inplace=True)
```

```python
import matplotlib.pyplot as plt

# Select numeric columns from merged_df
numeric_cols = merged_df.select_dtypes(include='number').columns

# Loop through columns and remove outliers
for col in numeric_cols:
    Q1 = merged_df[col].quantile(0.25)
    Q3 = merged_df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Filter to remove outliers
    filtered_data = merged_df[(merged_df[col] >= lower_bound) & (merged_df[col] <= upper_bound)]

    # Plot after removing outliers
    plt.figure(figsize=(6, 3))
    plt.boxplot(filtered_data[col].dropna(), vert=False)
    plt.title(f'Box Plot for {col} (Outliers Removed)')
    plt.xlabel(col)
    plt.tight_layout()
    plt.show()
```

# Clustering Analysis: Identifying High-Risk Groups Using K-Means

This section applies K-Means clustering to segment population groups based on:
- Hospitalisation rates
- Risk factors
- Gender
- Age group

Categorical variables are label-encoded, and all features are standardized.

```python
# Import libraries
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
#Copy your dataset
cluster_df = merged_df.copy()

# Label encode categorical columns
label_cols = ['Risk_Factor', 'Sex', 'Age_Group']
for col in label_cols:
    cluster_df[col] = LabelEncoder().fit_transform(cluster_df[col].astype(str))

#Select required features
features = ['Hospitalisations_per_100k', 'Risk_Factor', 'Sex', 'Age_Group']
X = cluster_df[features]
```

```python
#Drop missing values
X = X.dropna()
cluster_df = cluster_df.loc[X.index]  # Keep matching rows only

#Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
cluster_df['Cluster'] = kmeans.fit_predict(X_scaled)

#Reduce dimensions with PCA for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
#Plot the clusters
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=cluster_df['Cluster'], cmap='viridis', alpha=0.6)
plt.title("KMeans Clustering Visualized with PCA")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.grid(True)
plt.tight_layout()
plt.show()
```
