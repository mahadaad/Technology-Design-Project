### 1.3 PHN_HBP_SEX_2017-18

```python
# Load and preview the 'PHA_CVD_DEATH_SEX' sheet
df2 = xls.parse('1.3 PHN_HBP_SEX_2017-18', skiprows=9)
df2.head()
```

```python
# Load with multi-level header (first 2 rows after skipping 9)
df2_multi = xls.parse('1.3 PHN_HBP_SEX_2017-18', skiprows=9, header=[0, 1])

# Flatten multi-level headers: Combine both header rows into a single line with clear labels
df2_multi.columns = [
    f"{str(col1).strip()} - {str(col2).strip()}" if str(col2).strip() != 'nan' else str(col1).strip()
    for col1, col2 in df2_multi.columns
]

# Remove rows where the 'Code' column is NaN (region code column will contain 'Code' string)
df2_multi = df2_multi[df2_multi[df2_multi.columns[0]] != 'Code']
df2_multi = df2_multi[df2_multi[df2_multi.columns[0]].notna()]

# Reset index for clarity
df2_multi.reset_index(drop=True, inplace=True)

for col in df2_multi.columns:
    print(repr(col))
```

```python
import pandas as pd

# Step 1: Rename key columns for simplicity
df2_multi.rename(columns={
    df2_multi.columns[0]: "Code",
    df2_multi.columns[1]: "Region"  
}, inplace=True)

# Step 2: Clean column names
df2_multi.columns = df2_multi.columns.str.strip()

# Step 3: Select only relevant columns
selected_columns = [
    "Code", "Region",
    'Persons aged 18+ - Number',
    'Persons aged 18+ - Proportion (%)',
    'Persons aged 18+ - 95% CI of proportion',
    'Persons aged 18+ - Population',
    'Persons aged 18+ - Age-standardised rate (per 100)',
    'Persons aged 18+ - 95% CI standardised rate',
    'Persons aged 18+ - Rate ratio  (relative to all of Australia)',
    'Persons aged 18+ - 95% CI rate ratio',
    'Persons aged 18+ - 95% CI rate ratio.1',
    'Males aged 18+ - Number',
    'Males aged 18+ - Proportion (%)',
    'Males aged 18+ - 95% CI of proportion',
    'Males aged 18+ - Population',
    'Males aged 18+ - Age-standardised rate (per 100)',
    'Males aged 18+ - 95% CI standardised rate',
    'Males aged 18+ - Rate ratio  (relative to all of Australia)',
    'Males aged 18+ - 95% CI rate ratio',
    'Males aged 18+ - 95% CI rate ratio.1',
    'Females aged 18+ - Number',
    'Females aged 18+ - Proportion (%)',
    'Females aged 18+ - 95% CI of proportion',
    'Females aged 18+ - Population',
    'Females aged 18+ - Age-standardised rate (per 100)',
    'Females aged 18+ - 95% CI standardised rate',
    'Females aged 18+ - Rate ratio  (relative to all of Australia)',
    'Females aged 18+ - 95% CI rate ratio'
]

# Select available columns safely
existing_columns = [col for col in selected_columns if col in df2_multi.columns]
df2_selected = df2_multi[selected_columns].copy()

# Step 4: Melt to long format
df2_final = pd.melt(
    df2_selected,
    id_vars=["Code", "Region"],
    var_name="Group_Metric",
    value_name="Value"
)

# Step 6: Assign Gender and AgeGroup, clean Metric labels
df2_final["Gender"] = df2_final["Group_Metric"].str.extract(r'^(Persons|Males|Females)')
df2_final["AgeGroup"] = df2_final["Group_Metric"].str.extract(r'aged (.*) -')
df2_final["Metric"] = df2_final["Group_Metric"].str.extract(r'- (.*)$')
df2_final.drop(columns="Group_Metric", inplace=True)

# Step 7: Standardize metric names
df2_final["Metric"] = df2_final["Metric"].replace({
    "95% CI of proportion": "CI_Proportion",
    "95% CI standardised rate": "CI_Standardised_Rate",
    "95% CI rate ratio": "CI_Rate_Ratio",
    "Proportion (%)": "Proportion",
    "Rate ratio  (relative to all of Australia)": "Rate_Ratio",  # double space!
    "Age-standardised rate (per 100)": "Age_Standardised_Rate"
})

# Step 8: Pivot to wide format
df_wide = df2_final.pivot_table(
    index=["Code", "Region", "Gender", "AgeGroup"],
    columns="Metric",
    values="Value",
    aggfunc="first"
).reset_index()

# Step 9: Split confidence intervals into lower/upper columns
def split_ci(ci_value):
    if isinstance(ci_value, str) and '–' in ci_value:
        parts = ci_value.strip("()").split("–")
        try:
            return float(parts[0]), float(parts[1])
        except:
            return None, None
    return None, None

for ci_col in ['CI_Proportion', 'CI_Standardised_Rate', 'CI_Rate_Ratio']:
    df_wide[f"{ci_col}_Lower"], df_wide[f"{ci_col}_Upper"] = zip(*df_wide[ci_col].apply(split_ci))
    df_wide.drop(columns=ci_col, inplace=True)

# Step 10: Convert proportion to percent format (round to 2 decimals)
df_wide["Proportion (%)"] = pd.to_numeric(df_wide["Proportion"], errors="coerce").round(2)
df_wide.drop(columns="Proportion", inplace=True)

# Step 11: Convert remaining metrics to numeric
numeric_cols = [col for col in df_wide.columns if col not in ["Code", "Region", "Gender", "AgeGroup"]]
for col in numeric_cols:
    df_wide[col] = pd.to_numeric(df_wide[col], errors="coerce")

# Cleaning Steps
# Step 12: Drop rows with missing standardised rate info and copy
hbp_df_clean = df_wide.dropna(subset=['Age_Standardised_Rate']).copy()

# Step 13: Standardize Region names (safe assignment)
hbp_df_clean.loc[:, 'Region'] = hbp_df_clean['Region'].str.strip().str.upper()

# Step 14: Reset index
hbp_df_clean.reset_index(drop=True, inplace=True)

# Step 14.1: Add Risk_Factor column
hbp_df_clean["Risk_Factor"] = "High blood pressure"

# Step 15: Confirm cleaning
print("Missing values after cleaning:\n", hbp_df_clean.isnull().sum())
print(" Final shape:", hbp_df_clean.shape)

# Step 11: Export to CSV
hbp_df_clean.to_csv("PHN_HBP_SEX_2017-18.csv", index=False, encoding='utf-8-sig')
print("Saved as 'PHN_HBP_SEX_2017-18.csv' successfully!")
```

```python
import pandas as pd

# Step 1: Define file paths
file_paths = {
    "PHN_CS_AGE_SEX": "PHN_CS_AGE&SEX.csv",
    "PHN_OB_AGE_SEX": "PHN_OB_AGE_SEX.csv",
    "PHN_IPA_AGE_SEX": "PHN_IPA_AGE_SEX.csv",
    "PHN_HBP_AGE_SEX": "PHN_HBP_AGE_SEX.csv",
}

# Step 2: Load datasets
dataframes = {name: pd.read_csv(path) for name, path in file_paths.items()}

# Step 3: Get union of all columns
all_columns = sorted(set().union(*(df.columns for df in dataframes.values())))

# Step 4: Standardize all DataFrames to the same columns
aligned_dfs = []
for name, df in dataframes.items():
    aligned_df = df.copy()
    for col in all_columns:
        if col not in aligned_df.columns:
            aligned_df[col] = pd.NA
    aligned_df = aligned_df[all_columns]  # reorder
    aligned_df["Source_File"] = name
    aligned_dfs.append(aligned_df)

# Step 5: Strictly filter out DataFrames that are all-NA by row or column
non_empty_dfs = []
for name, df in zip(file_paths.keys(), aligned_dfs):
    # Check: at least one non-NA row and one non-NA column
    if df.dropna(how="all").shape[0] > 0 and df.dropna(axis=1, how="all").shape[1] > 0:
        non_empty_dfs.append(df)
    else:
        print(f"Skipped: {name} (empty or all-NA content)")

# Step 6: Concatenate without warning
if non_empty_dfs:
    merged_df = pd.concat(non_empty_dfs, ignore_index=True)
# Add Year column
    merged_df["Year"] = "2017–2018"

    merged_df.to_csv("Risk_Factors.csv", index=False, encoding="utf-8-sig")
    print(" Merged file saved as 'Risk_Factors.csv'")
else:
    print(" No valid data to merge.")
```
